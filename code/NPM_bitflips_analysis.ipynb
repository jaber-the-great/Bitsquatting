```python
#Checks whether the input strings are one bit away from each other or not
#     if cnt ==1:
#         print(x,y,"one bit away")
#     if cnt > 1:
#         print(x,y,"very far")
#     if cnt == 0: 
#         print(x,y,"equal")

def FindBitFlip(x,y):
    cnt = 0
    ValuesForOneBitDistance = [1,2,4,8,16,32,64,128,256]
    for i in range(len(x)):
            temp = ord(x[i]) ^ ord(y[i])
            if temp == 0:
                continue
            elif temp in ValuesForOneBitDistance:
                cnt+=1 
                if cnt> 1:
                    break
            else:
                cnt = temp
                break
    return cnt


```

```python
#Works for all types of input file, considers equal length words before checking for a flip 
def SearchFlips():
    file = open("names.txt","r")
    data = file.readlines()
    for i in range(len(data)):
        data[i] = data[i].strip('\n')

    cnt2 = 0
    cnt3 = 0
    for i in range(len(data)):

        pname = ' '.join(format(ord(i), '08b') for i in data[i])
        equals = [data[i]]
        bitflips = [data[i]]
        length = len(data[i])

        for j in range(len(data)):
            cnt3+=1
            if(len(data[j]) ==length):

                sname = ' '.join(format(ord(i), '08b') for i in data[j])
                result = FindBitFlip(pname,sname)
                if result == 1:
                    bitflips.append(data[j])
                if result == 0:
                    cnt+=1
        if len(bitflips) > 1:
            out = open(str(i)+"bitflips.txt","w")
            for element in bitflips:
                out.write(element)

        out.close()
        if cnt>1: 
            print(cnt)
        cnt = 0
```

```python
# Find the search domain for bitflip

def findIndex(length, ranges ):
    for i in range(len(ranges)):
        if length == ranges[i][1]:
            begin = ranges[i][0]
            end = ranges[i+1][0]
            break
    return [begin,end]
```
