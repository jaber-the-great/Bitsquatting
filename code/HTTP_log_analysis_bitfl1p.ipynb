```python
# Checks whether the given host name(in the log file) contains a legit unflipped domain name or not
def findInUnflipped(domain_list,name):
    for item in domain_list:
        if item in name:
            return True
    return False
```

```python
# Checks whether the given host name(in the log file) contains a flipped domain name or not
def findInFlipped(domain_list,name):
    for item in domain_list:
        for value in domain_list[item]:
            if value in name :
                return True
    return False
```


If the query is unflipped domain, DNS server responds with: (unflipped, 168.235.68.44 )

If the query is flipped domain, the DNS server responds with 2 answers:
(unflipped, 168.235.68.44) and (flipped, 168.235.68.45)

**Therefore, we have 4 different combinations**
- [ ] Check later
(we have one other type of log which the hostname is not valid or it is an IP address.)

```python
# case1 and 2: unflipped domain with corresponding IP address and the other address

corresponding = open("unflipped_corresponding.txt","w")
NON_corresponding = open("unflipped_NON_corresponding.txt","w")
erronous = open("unflipped_no_desIP.txt","w")

for i in range(len(data)):
    hostname = data[i]['httpHost']
    if findInUnflipped(domains,hostname):        
        if data[i]['dst'] == '168.235.68.44':
            corresponding.write(json.dumps(data[i])+'\n')
        elif data[i]['dst'] == '168.235.68.45':
            NON_corresponding.write(json.dumps(data[i])+ '\n')
        else:
            erronous.write(json.dumps(data[i])+ '\n')
corresponding.close()
NON_corresponding.close()
erronous.close()
```

```python
# Find and print the useragents of case1 and case2
case1agent = set()
case2agent = set()
error = set()
for i in range(len(data)):
    hostname = data[i]['httpHost']
    if findInUnflipped(domains,hostname):        
        if data[i]['dst'] == '168.235.68.44':
            case1agent.add(data[i]['userAgent'])
        elif data[i]['dst'] == '168.235.68.45':
            case2agent.add(data[i]['userAgent'])
        else:
            error.add(data[i]['userAgent'])

```

```python
# case3 and 4: flipped domain with corresponding IP address and the other address

corresponding = open("flipped_corresponding.txt","w")
NON_corresponding = open("flipped_NON_corresponding.txt","w")
erronous = open("flipped_no_desIP.txt","w")

for i in range(len(data)):
    hostname = data[i]['httpHost']
    if findInFlipped(domains,hostname):        
        if data[i]['dst'] == '168.235.68.44':
            corresponding.write(json.dumps(data[i])+'\n')
        elif data[i]['dst'] == '168.235.68.45':
            NON_corresponding.write(json.dumps(data[i])+ '\n')
        else:
            erronous.write(json.dumps(data[i])+ '\n')
corresponding.close()
NON_corresponding.close()
erronous.close()
```
